### モデルデータ作成

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

### 説明変数とターゲット変数をわける

```python
X = np.array(dataset_new.drop('y',axis=1))
y = np.array(dataset_new[['y']])
print(np.sum(y==1),np.sum(y==0))
```

### 不均衡データをUnderSampling

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
sampler_under = RandomUnderSampler(random_state=42)
X_under,y_under = sampler_under.fit_sample(X,y)
print(X_under.shape,y_under.shape)
print(np.sum(y_under==1),np.sum(y_under==0))
```

### 学習データを70%(X_train, y_train)と30%(X_train, y_train)に分割

```python
X_under_train, X_under_test, y_under_train, y_under_test = train_test_split(X_under, y_under, test_size=0.3, random_state=1234)
```

```python
print(X_under_train.shape,y_under_train.shape)
```

```python
print(X_under_test.shape,y_under_test.shape)
```

# ランダムフォレストでモデル作成

```python
rf = RandomForestClassifier()
```

```python
params = {'n_estimators': [100,750], 'max_depth': [20,50]}
```

ハイパーパラメータのチューニングをGridSearchで行う

```python
gscv = GridSearchCV(rf, param_grid=params, verbose=1,cv=3)
```

```python
gscv.fit(X_under_train, y_under_train)
```

```python
print(gscv.best_score_)
print(gscv.best_params_)
```

### ハイパーパラメータチューニング後のスコアを見てみる

```python
rf = RandomForestClassifier(n_estimators=20, max_depth=750)
rf.fit(X_under_train, y_under_train)

print(rf.predict(X_under_train))
print(rf.score(X_under_train,y_under_train))
print(rf.score(X_under_test,y_under_test))
```

```python
fti = rf.feature_importances_   
dataset_new2 = dataset_new.drop('y', axis=1)
importance =[]
print('Feature Importances:')
for i, feat in enumerate(dataset_new2.columns):
    importance.append(fti[i])
pd.Series(importance, index = dataset_new2.columns).sort_values(ascending = False)
```

```python
from sklearn.metrics import confusion_matrix
X_under_train_pred=rf.predict(X_under_train)
X_under_train_pred
print(confusion_matrix(y_under_train,X_under_train_pred))

X_under_test_pred=rf.predict(X_under_test)
print(confusion_matrix(y_under_test,X_under_test_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_under_test,X_under_test_pred))
```
